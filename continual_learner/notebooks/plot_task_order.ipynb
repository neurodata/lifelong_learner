```python
import pickle
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
import numpy as np
from itertools import product
import seaborn as sns
sns.set()

%matplotlib inline

def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
eta = np.zeros((10,10))
file_pos = np.asarray(range(1,11))

for i in range(10):
    file_pos = np.roll(file_pos,1)
    for j in range(10):
        filename = './order_res/EWC'+str(file_pos[j])
        print('loading '+filename)
        err = unpickle(filename)
        print(err[j,j],err[9,j])
        eta[i][j] = err[j,j]/err[9,j]
```

```python
eta
```

```python
position = ['pos 1','pos 2','pos 3','pos 4','pos 5','pos 6','pos 7','pos 8','pos 9','pos 10']
fig, ax = plt.subplots(1,1, figsize=(10,8))
clr = ["#e41a1c", "#a65628", "#377eb8", "#4daf4a", "#984ea3", "#ff7f00", "#000000"]
c = sns.color_palette(clr, n_colors=10)
fontsize = 26
for i in range(10):
    for j in range(10):
        if i == 0:
            ax.scatter(i+1,eta[i,j],color=c[j],label=position[j])
        else:
            ax.scatter(i+1,eta[i,j],color=c[j])
    
ax.set_xticks(np.arange(1,11))
ax.tick_params(labelsize=20)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=22)
ax.set_title('EWC',fontsize=28)
ax.set_xlabel('task number',fontsize=fontsize)
ax.set_ylabel('Final BTE',fontsize=fontsize)
plt.savefig('cifar-100-order.png')
```
